```python
import tkinter as tk
from tkinter import ttk
from tkinter import *

window = tk.Tk()
window.title("Election Software")
window.geometry("500x500")

def remove_text(x):
    x.delete(0,"end")

def tab1():

    frame1= ttk.Frame(window)
    frame1.pack()

    heading= ttk.Label(frame1, text="Welcome to the elections", font="Calibri 20")
    heading.pack()

    info=tk.Label(frame1,text="Hello and welcome to the elections\nPls follow the rules\nOr you will die", font="Calibri 20")
    info.pack(pady=100)

    def next_func():
        frame1.destroy()
        tab2()
    
    next = ttk.Button(frame1, text="I Understand", command=next_func)
    next.pack(pady=50)

def tab2():

    frame2=ttk.Frame(window)
    frame2.pack()

    heading= ttk.Label(frame2, text="LOGIN", font="Calibri 20")
    heading.pack()

    def remove_username(e):
        username.delete(0,"end")

    username= ttk.Entry(frame2)
    username.insert(0, "USERNAME")
    username.bind("<FocusIn>",remove_username)
    username.pack(pady=100)

    def remove_password(e):
        password.delete(0,"end")

    password= ttk.Entry(frame2)
    password.insert(0, "PASSWORD")
    password.bind("<FocusIn>",remove_password)
    password.pack()

    def submit_func():
        frame2.destroy()
        tab3()

    submit= ttk.Button(frame2, text="SUBMIT", command=submit_func)
    submit.pack(pady=50)
    
def tab3():

    frame3= ttk.Frame(window)
    frame3.pack()

    heading= ttk.Label(frame3, text="HEAD BOY", font="Calibri 20")
    heading.pack()


    








tab1()

window.mainloop()
```
